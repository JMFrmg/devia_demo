## Chargement et agrégation des données

```python
df_usagers = pd.read_csv("./data/usagers-2023.csv", sep=";")
df_vehicules = pd.read_csv("./data/vehicules-2023.csv", sep=";")
df_lieux = pd.read_csv("./data/lieux-2023.csv", sep=";")
df_caracteristiques = pd.read_csv("./data/caract-2023.csv", sep=";")
```

```python
df_usagers.head()
```

```python
df_vehicules.head()
```

```python
df_merged = df_usagers.merge(df_vehicules, on="id_vehicule", how="inner")
```

```python
df_lieux.head()
```

```python
df_merged.head()
```

```python
df_merged = df_merged.merge(df_lieux, left_on="Num_Acc_x", right_on="Num_Acc", how="inner")
```

```python
df_caracteristiques.head()
```

```python
df = df_merged.merge(df_caracteristiques, left_on="Num_Acc_x", right_on="Num_Acc", how="inner", suffixes=("_xy", "_yx"))
```
